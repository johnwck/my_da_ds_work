## **Step 1: Read documents from csv file** 

```python
import pandas as pd
```

```python
corpus = pd.read_csv('data2.csv', encoding='cp1252')
```

```python
pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_columns', None)
```

```python
corpus
```

<br>

## **Step 2: Remove non alphanumeric characters in documents**

```python
corpus['document_cleaned'] = corpus['document'].replace('[^A-Za-z0-9]', ' ', regex=True)
```

```python
corpus
```

<br>

## **Step 3: Change characters in documents to lower case**

```python
corpus['document_cleaned'] = corpus['document_cleaned'].str.lower()
```

```python
corpus
```

<br>

## **Step 4: Get stopwords from NLTK**

```python
from nltk.corpus import stopwords
```

```python
print(stopwords.words('english'))
```

<br>

## **Step 5: Split documents from Step 3 and Step 4 into list of tokens (words) without stopwords**

```python
corpus['token_list'] = corpus['document_cleaned'].apply(lambda x: [word for word in x.split() if word not in stopwords.words('english')])
```

```python
corpus
```

<br>

## **Step 6: Count each and every token from Step 5**

```python
from collections import defaultdict
```

```python
token_frequency = defaultdict(int)

for each_token_list in corpus['token_list']:
    for token in each_token_list:
        token_frequency[token] = token_frequency[token] + 1
```

<br>

## **Step 7: Form list of tokens with more than count of 1 from Step 6**

```python
corpus['token_list_new'] = corpus['token_list'].apply(lambda x: [i for i in x if token_frequency[i]>1])
```

```python
corpus
```

<br>

## **Step 8: Put unique tokens from Step 7 in dictionary**

```python
from gensim import corpora
```

```python
dictionary_unique_tokens = corpora.Dictionary(corpus['token_list_new'])
```

```python
print(dictionary_unique_tokens)
```

<br>

## **Step 9: Convert documents from Step 7 to bag of words vector using dictionary in Step 8**

```python
corpus['bag_of_words_vector'] = corpus['token_list_new'].apply(lambda x: dictionary_unique_tokens.doc2bow(x))
```

```python
corpus
```

<br>

## **Others: Get id of unique tokens**

```python
print(dictionary_unique_tokens.token2id)
```

**This is the X in (X,Y) bag of words token**

<br>

## **Others: Get frequency of unique tokens**

```python
# sorted by token

print(sorted(token_frequency.items()))
```

```python
# sorted by token frequency

print(sorted(token_frequency.items(), key=lambda item: item[1]))
```

<br>

## **Step 10: Training of model to 3 topics with highest coherence score**

```python
from gensim import models
```

```python
from gensim.models.coherencemodel import CoherenceModel
```

```python
import pprint
```

```python
lsi_model = models.LsiModel(corpus=corpus['bag_of_words_vector'], id2word=dictionary_unique_tokens, num_topics=3, random_seed=42, power_iters=1000, onepass=False)

cm = CoherenceModel(model=lsi_model, corpus=corpus['bag_of_words_vector'], coherence='u_mass')
coherence = cm.get_coherence()

print('----- 3 topics are -----')
pprint.pprint(lsi_model.print_topics())
print()

print('----- coherence score for 3 topics is -----')
print(coherence)
```

<br>

## **Step 11: Classification with trained model on existing documents**

```python
corpus_lsi = lsi_model[corpus['bag_of_words_vector']]
```

```python
df1 = []
for i in corpus_lsi:
    df1.append([i[0][1], i[1][1], i[2][1]])

df2 = pd.DataFrame(df1, columns=['topic_0_fashion','topic_1_data_science','topic_2_mental_illness'])
```

```python
df2['original_topic'] = corpus['topic']
```

## **Step 12: Classification with trained model on new documents**

<br>

**New documents**

```python
new_document_1 = 'Data Science is complicated due to statistics and mathematics.'
new_document_2 = 'Mental illness can affect behavior and mood.'
new_document_3 = 'Fashion can be defined by styles and clothes of people.'
```

```python
print(new_document_1)
print(new_document_2)
print(new_document_3)
```

<br>

**Remove non alphanumeric characters**

```python
import re
```

```python
new_document_1a = re.sub(r'[^A-Za-z0-9]', ' ', new_document_1)
new_document_2a = re.sub(r'[^A-Za-z0-9]', ' ', new_document_2)
new_document_3a = re.sub(r'[^A-Za-z0-9]', ' ', new_document_3)
```

```python
print(new_document_1a)
print(new_document_2a)
print(new_document_3a)
```

<br>

**Change to lower case**

```python
new_document_1a = new_document_1a.lower()
new_document_2a = new_document_2a.lower()
new_document_3a = new_document_3a.lower()
```

```python
print(new_document_1a)
print(new_document_2a)
print(new_document_3a)
```

<br>

**Form list of tokens without stopwords**

```python
new_document_1a = [word for word in new_document_1a.split() if word not in stopwords.words('english')]
new_document_2a = [word for word in new_document_2a.split() if word not in stopwords.words('english')]
new_document_3a = [word for word in new_document_3a.split() if word not in stopwords.words('english')]
```

```python
print(new_document_1a)
print(new_document_2a)
print(new_document_3a)
```

```python
new_document_1a = [word for word in new_document_1a if token_frequency[word]>1]
new_document_2a = [word for word in new_document_2a if token_frequency[word]>1]
new_document_3a = [word for word in new_document_3a if token_frequency[word]>1]
```

```python
print(new_document_1a)
print(new_document_2a)
print(new_document_3a)
```

<br>

**Convert to bag of words vector (use dictionary formed earlier)**

```python
new_document_1a = dictionary_unique_tokens.doc2bow(new_document_1a)
new_document_2a = dictionary_unique_tokens.doc2bow(new_document_2a)
new_document_3a = dictionary_unique_tokens.doc2bow(new_document_3a)
```

```python
print(new_document_1a)
print(new_document_2a)
print(new_document_3a)
```

<br>

**Classify new documents (use model trained earlier)**

```python
new_document_1_class = lsi_model[new_document_1a]
new_document_2_class = lsi_model[new_document_2a]
new_document_3_class = lsi_model[new_document_3a]
```

```python
print('topic_0_fashion:', round(new_document_1_class[0][1],4), '\t',
      'topic_1_data_science:', round(new_document_1_class[1][1],4), '\t',
      'topic_2_mental_illness:', round(new_document_1_class[2][1],4), '\t',
      new_document_1)
print('topic_0_fashion:', round(new_document_2_class[0][1],4), '\t',
      'topic_1_data_science:', round(new_document_2_class[1][1],4), '\t',
      'topic_2_mental_illness:', round(new_document_2_class[2][1],4), '\t',
      new_document_2)
print('topic_0_fashion:', round(new_document_3_class[0][1],4), '\t',
      'topic_1_data_science:', round(new_document_3_class[1][1],4), '\t',
      'topic_2_mental_illness:', round(new_document_3_class[2][1],4), '\t',
      new_document_3)
```
