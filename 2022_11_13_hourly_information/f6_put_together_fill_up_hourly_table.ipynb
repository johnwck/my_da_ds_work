# <font color=blue>Import  
    

```python
import pandas as pd
from datetime import datetime, timedelta
import seaborn as sns
import matplotlib.pyplot as plt
pd.set_option('display.max_columns', None)
```

<br>  

# <font color=blue>Create dataframe with start time and end time of jobs  
    

```python
df = {'job_name'       : [           'job_1',            'job_2',            'job_3',            'job_4',            'job_5',            'job_6'],
       'start_date_time': ['2022-11-20 22:00', '2022-11-20 22:30', '2022-11-20 23:00', '2022-11-20 23:15', '2022-11-20 23:45', '2022-11-21 01:45'],
       'end_date_time'  : ['2022-11-20 22:45', '2022-11-20 23:00', '2022-11-20 23:30', '2022-11-20 23:45', '2022-11-21 01:45', '2022-11-22 01:45']}
```

```python
job_table = pd.DataFrame()
```

```python
job_table['job_name'] = df['job_name']
job_table['start_date_time'] = df['start_date_time']
job_table['end_date_time'] = df['end_date_time']
```

```python
column_list = job_table.iloc[:, 1:3].columns
job_table[column_list] = job_table[column_list].astype('datetime64[ns]')
```

```python
job_table
```

```python
job_table.info()
```

<br>  

# <font color=blue>Correct end time due to overlap of start date time and end date time  
    

```python
# ========== Step 1 ==========
job_table.sort_values(by = ['start_date_time'], inplace = True)
job_table.reset_index(inplace = True, drop = True)

# ========== Step 2 ==========
j = len(job_table) - 1
for i in range(len(job_table)):
    if i != j:
        if (job_table.loc[i, 'start_date_time'] < job_table.loc[i + 1, 'start_date_time']) and \
           (job_table.loc[i + 1, 'start_date_time'] < job_table.loc[i, 'end_date_time']):
            job_table.loc[i, 'end_date_time'] = job_table.loc[i + 1, 'start_date_time']

job_table
```

<br>  

# <font color=blue>Split to next line for job with cross-over to next day  
    

```python
i = 0
exit_loop = False
while exit_loop == False:
    if job_table.loc[i, 'start_date_time'].date() != job_table.loc[i, 'end_date_time'].date():
        job_table.loc[i + 0.5, 'job_name'] = job_table.loc[i, 'job_name']
        job_table.loc[i + 0.5, 'start_date_time'] = datetime(job_table.loc[i, 'end_date_time'].year,
                                                                job_table.loc[i, 'end_date_time'].month,
                                                                job_table.loc[i, 'end_date_time'].day,
                                                                0, 0, 0)
        job_table.loc[i + 0.5, 'end_date_time'] = job_table.loc[i, 'end_date_time']
        job_table.loc[i, 'end_date_time'] = datetime(job_table.loc[i, 'end_date_time'].year,
                                                        job_table.loc[i, 'end_date_time'].month,
                                                        job_table.loc[i, 'end_date_time'].day,
                                                        0, 0, 0)
        job_table.sort_index(inplace = True)
        job_table.reset_index(inplace = True, drop = True)
        i = i + 2
    else:
        i = i + 1
    if i >= len(job_table):
        exit_loop = True
    else:
        exit_loop = False

job_table
```

# <font color=blue>Create hourly table  
    

```python
hourly_table = pd.DataFrame(columns=['job_name',
                                     'start_date_time', 'end_date_time',
                                     'hr_00', 'hr_01', 'hr_02', 'hr_03', 'hr_04', 'hr_05', 'hr_06', 'hr_07', 'hr_08',
                                     'hr_09', 'hr_10', 'hr_11', 'hr_12', 'hr_13', 'hr_14', 'hr_15', 'hr_16', 'hr_17',
                                     'hr_18', 'hr_19', 'hr_20', 'hr_21', 'hr_22', 'hr_23'])
```

```python
hourly_table['job_name'] = job_table['job_name'].astype(str)
```

```python
hourly_table['start_date_time'] = job_table['start_date_time']
hourly_table['end_date_time'] = job_table['end_date_time']
column_list = hourly_table.iloc[:, 1:3].columns
hourly_table[column_list] = hourly_table[column_list].astype('datetime64[ns]')
```

```python
column_list = hourly_table.iloc[:, 3:27].columns
hourly_table[column_list] = hourly_table[column_list].astype(float)
```

```python
hourly_table.fillna(0.0, inplace = True)
```

```python
hourly_table.info()
```

```python
hourly_table
```

<br>  

# <font color=blue>Fill up hourly table  
    

```python
for j in range(len(hourly_table)):
    
    for i in range(24):
    
        compare_hour = i
    
        compare_time = datetime(hourly_table.loc[j, 'start_date_time'].year,
                                hourly_table.loc[j, 'start_date_time'].month,
                                hourly_table.loc[j, 'start_date_time'].day,
                                compare_hour, 0, 0)

        start_hour = hourly_table.loc[j, 'start_date_time'].hour

        end_hour = hourly_table.loc[j, 'end_date_time'].hour
        end_minute = hourly_table.loc[j, 'end_date_time'].minute
        if end_hour == 0 and end_minute == 0:
            end_hour = 23
        else:
            if end_minute == 0:
                end_hour = end_hour - 1

        if start_hour == compare_hour:
            hourly_table.iloc[j, i + 3] = (min(hourly_table.loc[j, 'end_date_time'], compare_time + timedelta(hours = 1)) - \
                                           hourly_table.loc[j, 'start_date_time']).total_seconds() / 60 / 60
        else:
            if end_hour == compare_hour:
                hourly_table.iloc[j, i + 3] = (hourly_table.loc[j, 'end_date_time'] - compare_time).total_seconds() / 60 / 60
            else:
                if hourly_table.loc[j, 'start_date_time'] < compare_time and compare_time < hourly_table.loc[j, 'end_date_time']:
                    hourly_table.iloc[j, i + 3] = 1
                else:
                    hourly_table.iloc[j, i + 3] = 0

hourly_table
```

# <font color=blue>Create start_date and duration_hour  
    

```python
hourly_table['start_date'] = hourly_table['start_date_time'].dt.date
hourly_table['start_date'] = hourly_table['start_date'].astype('datetime64[ns]')
```

```python
hourly_table['duration_hour'] = hourly_table[['hr_00', 'hr_01', 'hr_02', 'hr_03', 'hr_04', 'hr_05', 'hr_06', 'hr_07',
                                              'hr_08', 'hr_09', 'hr_10', 'hr_11', 'hr_12', 'hr_13', 'hr_14', 'hr_15',
                                              'hr_16', 'hr_17', 'hr_18', 'hr_19', 'hr_20', 'hr_21', 'hr_22', 'hr_23']].sum(axis = 1)
```
